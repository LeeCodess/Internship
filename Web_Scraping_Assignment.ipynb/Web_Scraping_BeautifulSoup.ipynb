```python
# question 2 - write a python program to display imdb’s top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)

    movie_name = soup.find_all("td", class_ = "titleColumn")
    title = []
    for i in movie_name:
        i = i.text.replace('\n','')
        i = i.rstrip(i[-6:])
        head,sep,tail = i.partition('.')
        title.append(tail.strip())
    title = title[:100]

    year_of_release = soup.find_all("span", class_ = "secondaryInfo")
    year = []
    for i in year_of_release:
        year.append(i.text.replace('(','').replace(')', ''))
    year = year[:100]
    
    rating_of_the_movie = soup.find_all("td", class_ = "ratingColumn imdbRating")
    rating = []
    for i in rating_of_the_movie:
        rating.append(i.text.replace('\n',''))
    rating = rating[:100]

    data = pd.DataFrame({"Name of the Movie" : title, "Rating of the Movie" : rating, "Year of Realease" : year})
    return(data)
```

```python
# question 4 - write s python program to display list of respected former presidents of india (i.e. name , term of office) from https://presidentofindia.nic.in/former-presidents.htm
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)

    names_of_the_president = soup.find_all("div", class_ = "presidentListing")
    Name = []
    for i in names_of_the_president:
        i = i.text
        head,sep,tail = i.partition('(')
        Name.append(head.replace('\n',''))

    term_of_office = soup.find_all("div", class_ = "presidentListing")
    Term = []
    for i in term_of_office:
        i = i.text
        head,sep,tail = i.partition('Term of Office: ')
        i = tail
        head,sep,tail = i.partition('\n')
        Term.append(head)
        
    data = pd.DataFrame({"Name of the Indian President" : Name, "Term of Office" : Term})
    return(data)
```

```python
# question 7 - write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    
    title = soup.find_all('a', class_ ='LatestNews-headline')
    Title = []
    for i in title:
        Title.append(i.text) 

    time = soup.find_all('time', class_ ='LatestNews-timestamp')
    Time = []
    for i in time:
        Time.append(i.text) 

    link = soup.find_all('div', class_='LatestNews-headlineWrapper')
    Link = []
    for i in link:
        head, sep, tail = str(i).partition('" title=')
        head, sep, tail = head.partition('a class="LatestNews-headline" href="')
        Link.append(tail)

    data = pd.DataFrame({'Title of the News':Title,'Time of the News':Time,'Link of the News':Link})
    return(data)
```

```python
html_content('https://www.cnbc.com/world/?region=world')
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)

    title = soup.find_all('h2', class_ ='sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR')
    Title = []
    for i in title:
        Title.append(i.text) 

    author = soup.find_all('span', class_ ='sc-1w3fpd7-0 pgLAT')
    Author = []
    for i in author:
        Author.append(i.text) 

    date = soup.find_all('span', class_ ='sc-1thf9ly-2 bKddwo')
    Date = []
    for i in date:
        Date.append(i.text) 

    link = soup.find_all('a', class_ ='sc-5smygv-0 nrDZj')
    Link = []
    for i in link:
        Link.append(i['href'])
        
    data = pd.DataFrame({'Name of the Article':Title,'Name of the Author':Author,'Date of Publication':Date,'Link of the Article':Link})
    return(data)
```

```python
# question 9 - write a python program to scrape mentioned details from dineout.co.in
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)
    
    Hotel_name = []
    Location = []
    for i in soup.find_all('div', class_ = 'restnt-info cursor'):
        Hotel_name.append(i.a.text)
        Location.append(i.div.text)

    Cuisine = []
    for i in soup.find_all('span', class_ = 'double-line-ellipsis'):
        head, sep, tail = i.text.partition('|')
        Cuisine.append(tail)

    link = []
    for i in soup.find_all('img', class_ = 'no-img'):
        link.append(i['data-src'])

    Rating = []
    for i in soup.find_all('div', class_ = 'restnt-rating rating-4'):
        Rating.append(i.text)

    data = pd.DataFrame({'Name of the Hotel':Hotel_name, 'Cuisine of the Hotel':Cuisine, 'Location of the Hotel':Location, 'Rating of the Hotel':Rating, 'Link of the Hotel':link})
    return(data)       
```

```python
# question 10 - write a python program to scrape the details of top publications from google scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def html_content(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content)

    rank = soup.find_all('td', class_ ='gsc_mvt_p')
    Rank = []
    for i in rank:
        Rank.append(i.text) 

    publication = soup.find_all('td', class_ ='gsc_mvt_t')
    Publication = []
    for i in publication:
        Publication.append(i.text) 

    h5_index = soup.find_all('a', class_ ='gs_ibl gsc_mp_anchor')
    H5_index = []
    for i in h5_index:
        H5_index.append(i.text) 

    h5_median = soup.find_all('span', class_ ='gs_ibl gsc_mp_anchor')
    H5_median = []
    for i in h5_median:
        H5_median.append(i.text)
                                                        
    data = pd.DataFrame({'Rank':Rank,'Publication':Publication,'h5_index':H5_index,'h5_median':H5_median})
    return(data)
```
